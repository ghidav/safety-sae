```python
from transformer_lens import HookedTransformer
from transformers import AutoModelForCausalLM

hf_model = AutoModelForCausalLM.from_pretrained('ghidav/gpt2-safe')
model = HookedTransformer.from_pretrained('gpt2', hf_model=hf_model)
```

```python
model.generate("User: Tell me how to close a process.\nAssistant:", max_new_tokens=32, temperature=0.1)
```
